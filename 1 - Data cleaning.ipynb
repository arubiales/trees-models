## 2. Cleaning

```python
def cleaning_function(df):
#     df_column = df_column.astype(str)
    '''
    A function to clean the Football dataset
    '''
    if re.findall('€\d*', df) and 'M' in df: #replace million for a real value
        return float(re.sub('[^0-9\.]', '', df))*1000000
    elif re.findall('€\d*', df) and'K' in df: #replace K for a real value
        return float(re.sub('[^0-9\.]', '', df))*100000
    elif 'Left' in df: #Transform to categorical
        return 0
    elif 'Right' in df: #Transform to categorical
        return 1
    elif re.findall('\+\d*', df): #transform column non numeric with "+" to numeric
        first_number = int(re.findall('\d*', df)[0])
        second_number = int(re.findall('\d*', df)[2])
        return first_number+second_number
    elif re.findall("\d'\d", df):
        return float(re.sub("'", '.', df))
    elif re.findall("\d*lbs", df):
        return int(re.sub("lbs", '', df))
    else: 
        return df
```

```python
df_footbal = df_footbal.astype(str)
df_footbal_clean = df_footbal.applymap(cleaning_function)
df_footbal_clean
```

```python
for col in df_footbal_clean.columns:
    try:
        df_footbal_clean[col] = df_footbal_clean[col].astype(float)
    except:
        pass
```

```python
df_footbal_clean.dropna(inplace=True)
df_footbal_clean.drop_duplicates(inplace=True)
df_footbal_clean.reset_index(inplace=True, drop=True)
df_footbal_clean
```
